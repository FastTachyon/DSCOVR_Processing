## Calculate average energy

Get average and median of energy distributions (spectra)

```python
# ref. https://stackoverflow.com/a/50786849

A =  df.iloc[:, 4:54].fillna(0).values
bins = np.arange(A.shape[1]+1)
mids = 0.5*(bins[1:] + bins[:-1])

avgs = np.empty(A.shape[0])
vars = np.empty(A.shape[0])
for i, n in enumerate(A):
    try:
        avgs[i] = np.average(mids, weights=n)
        vars[i] = np.average((mids - avgs[i])**2, weights=n)
    except ZeroDivisionError:
        avgs[i] = np.nan
        vars[i] = np.nan
```

```python
df["Eavg"] = avgs
df["Evar"] = vars
```

## Check cross-correlation (lag) between signals

```python
b, a = signal.butter(4, 0.008)
tdf_i["Eavgff"] = signal.filtfilt(b, a, tdf_i["Eavg"], padlen=150)
tdf_i["Kp10ff"] = signal.filtfilt(b, a, tdf_i["Kp10"], padlen=150)
```

```python
tdf_i_subset = tdf_i[tdf_i.index>datetime.datetime(2022,2,2)].head(5000)
tdf_i_subset
```

```python
fig, ax = plt.subplots(figsize=(16,9))
tdf_i_subset.plot(y=["Kp10","Eavg", "Kp10ff", "Eavgff"], ax=ax)
plt.show()
```

```python
#ref: https://stackoverflow.com/a/56432463

def lag_finder(y1, y2, ts):
    n = len(y1)

    corr = signal.correlate(y2, y1, mode='same') / np.sqrt(signal.correlate(y1, y1, mode='same')[int(n/2)] * signal.correlate(y2, y2, mode='same')[int(n/2)])

    delay_arr = np.linspace(-0.5*n*ts, 0.5*n*ts, n)
    delay = delay_arr[np.argmax(corr)]
    print('y2 is ' + str(delay) + ' behind y1')

    plt.figure()
    plt.plot(delay_arr, corr)
    plt.title('Lag: ' + str(np.round(delay, 3)) + ' s')
    plt.xlabel('Lag')
    plt.ylabel('Correlation coeff')
    plt.show()

    return delay
```

```python
sample_time = tdf_i_subset.index.to_series().diff().median().seconds
sample_time
```

```python
lag_finder(tdf_i_subset.Eavg.values, tdf_i_subset.Kp10.values, sample_time)/3600
```
